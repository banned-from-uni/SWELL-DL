### 1D CNN

```python
feature_names = anova.index
```

```python
feature_names
```

```python
X = swell[feature_names].to_numpy()
#X = swell[feature_names[:15]].to_numpy()
y = swell['condition']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
print(y_train.unique())
train_class_names = y_train
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)
print(np.unique(y_train))
```

```python
print(type(X_train))
print(type(y_train))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train.shape)
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])
print(X_train.shape)
```

```python
from timeit import default_timer as timer

class TimingCallback(tf.keras.callbacks.Callback):
    def __init__(self, logs={}):
        self.logs=[]
    def on_epoch_begin(self, epoch, logs={}):
        self.starttime = timer()
    def on_epoch_end(self, epoch, logs={}):
        self.logs.append(timer()-self.starttime)

cb = TimingCallback()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='accuracy', mode='max', verbose=1, patience=10)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, MaxPooling1D

def create_model():
  model = Sequential()
  model.add(Conv1D(filters=64, kernel_size=2,
                   activation="relu", padding="same"))
  model.add(Dense(16, activation="relu"))
  model.add(MaxPooling1D(padding='same'))
  model.add(Flatten())
  model.add(Dense(3, activation = 'softmax'))

  model.compile(loss = 'sparse_categorical_crossentropy',
                optimizer = "adam",
                metrics = ['accuracy'])

  return model
```

```python
try:
  with tf.device('/CPU:0'):
    model = create_model()
    history = model.fit(X_train, y_train, batch_size=64,
              epochs=600, verbose = 1,
              validation_data=(X_test, y_test),
              callbacks = [early_stop,cb])
  print(f'Training timer: {sum(cb.logs)} seconds')
  cb.logs.clear()
except KeyboardInterrupt:
  cb.logs.clear()
  raise Exception('KeyboardInterrupt. Cleared timing logs.')
```

```python
loss, accuracy = model.evaluate(X_train, y_train)
print("Loss:", loss, " Accuracy:", accuracy)
```

```python
pred = model.predict(X_test)
pred_y = pred.argmax(axis=-1)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred_y)
```

```python
# old cm plot
print(cm)
plt.figure(figsize=(12,8),dpi=300)
sns.heatmap(cm, annot=True, cbar = False,fmt='.10g', annot_kws={"fontsize":30});
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_y))
```
